```python
def printhex(val):
    """Little helper function to print decimal in a hex
    """
    print(format(val, '02x'))
    
def printcharhex(ch):
    """Little helper to print out an ascii character in hex
    """
    printhex(ord(ch))
    
```

```python
def xorarray (arr1, arr2):
    """Helper function that xors two numpy arrays together
    """
    len1 = len(arr1)
    len2 = len(arr2)
    minlen = min(len1,len2)
    return arr1[:minlen].__xor__(arr2[:minlen])
```

```python
xorarray(npm[0],npm[1])
```

```python
# Build up an array of all of the messages xored with themselves 
# This gives us a frequency distribution 

onearray = np.array([0,0])
for m in npm:
    for n in npm:
        if not(np.array_equiv(m, n)):
            b = xorarray(m,n)
            onearray = np.concatenate((onearray, b))
 
# flatten the array 
onearray.flatten()

# count out the unique values
unique, counts = np.unique(onearray, return_counts=True)
frequencies = dict(zip(unique, counts))

# sort the array 
import operator
sorted_d = sorted(frequencies.items(), key=operator.itemgetter(1))

# Print out the most common 10 items
print ("ASCII, Count")
for i in sorted_d[-10:-1]:
    print (format(i[0], '02x'), ',', i[1])
```


# ASCII Manipulations

Uppercase letters A=0x41, Z=0x5A
Lowercase letters a=0x61, z=0x7a
Space = 0x20, Period = 0x2e, Comma = 0x2c, Quote = 0x22
The only way to get above 0x3f is to have a letter xor'ed with space, comma, quote or period 

~~~~~~~~~~~~~~~~
0x41 = 0100 0001
0x5a = 0101 1010
0x61 = 0110 0001
0x7a = 0111 1010

0x20 = 0010 0000
~~~~~~~~~~~~~~~~

No way to XOR a letter and another letter and get bit 6 set 

Xoring with space (0x20) yields opposite case


```python

target_sweep = len(smessage + 1)
space_array = np.zeros([10,target_sweep])

for i in range(0,10):
    print ("*")
    for j in range(i+1,10):        
        r = xorarray(npm[i], npm[j])
        count = 0
        for k in range(0,target_sweep):          
            if r[k] == 0x0:
                print (i,j,k, "equal")
            if r[k] >= 0x41 and r[k] <= 0x5a:
                print (i,j,k, "possible space")
                printhex(r[k])
                space_array[i,k] += 1
                space_array[j,k] += 1
            if r[k] >= 0x61 and r[k] <= 0x7a:
                print (i,j,k, "cap possible space")
                printhex(r[k])
                space_array[i,k] += 1
                space_array[j,k] += 1
            if r[k] <= 0x40 and r[k] != 0x0:
                print (i,j,k, "letter")
                space_array[i,k] -= 1
                space_array[j,k] -= 1

print(space_array)
```

```python

def use_spaces_to_extract_keys(threshold = 0x6):
    secretkey = np.zeros([target_sweep], dtype=np.uint8)

    collisions = 0

    for i in range(0,10):
        for j in range(0,target_sweep):
            if space_array[i,j] >= 0x9:
                if secretkey[j] != 0 and (npm[i][j] ^ 0x20 != secretkey[j]):
                    print("HMMM was", secretkey[j], "Setting to ", npm[i][j] ^ 0x20)
                    collisions += 1
                secretkey[j] = npm[i][j] ^ 0x20
            else:
                if space_array[i,j] >= threshold:
                    if secretkey[j] != 0 and (npm[i][j] ^ 0x20 != secretkey[j]):
                        print("HMMM was", secretkey[j], "Setting to ", npm[i][j] ^ 0x20)
                        collisions += 1
                    secretkey[j] = npm[i][j] ^ 0x20

    print ("number of collisions", collisions)
    return secretkey, collisions
```
